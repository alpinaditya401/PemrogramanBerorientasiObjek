```python
class Hewan:
    def __init__(self, nama, umur):
        self.nama = nama
        self.umur = umur
    def info(self):
        print(f"Nama: {self.nama}")
        print(f"Umur: {self.umur} tahun")
class Kucing(Hewan):
    def __init__(self, nama, umur, warna_bulu, ras):
        super().__init__(nama, umur)
        self.warna_bulu = warna_bulu
        self.ras = ras
    def info(self):
        print("=== INFO KUCING ===")
        super().info()
        print(f"Warna Bulu: {self.warna_bulu}")
        print(f"Ras: {self.ras}\n")
class Paus(Hewan):
    def __init__(self, nama, umur, habitat, berat_ton):
        super().__init__(nama, umur)
        self.habitat = habitat
        self.berat_ton = berat_ton

    def info(self):
        print("=== INFO PAUS ===")
        super().info()
        print(f"Habitat: {self.habitat}")
        print(f"Berat: {self.berat_ton} Ton\n")

class Singa(Hewan):
    def __init__(self, nama, umur, status_kawanan, kecepatan_lari):
        super().__init__(nama, umur)
        self.status_kawanan = status_kawanan
        self.kecepatan_lari = kecepatan_lari
    def info(self):
        print("=== INFO SINGA ===")
        super().info()
        print(f"Status Kawanan: {self.status_kawanan}")
        print(f"Kecepatan Lari: {self.kecepatan_lari} km/jam\n")
kucing_saya = Kucing("Milo", 2, "Oranye", "Anggora")
paus_biru = Paus("Willy", 45, "Samudra Pasifik", 150)
singa_afrika = Singa("Simba", 7, "Pemimpin (Alpha)", 80)


kucing_saya.info()
paus_biru.info()
singa_afrika.info()
```

```python
class Produk:
    def __init__(self, nama, harga):
        self.nama = nama
        self.harga = harga
    def info(self):
        print(f"Nama: {self.nama}")
        print(f"Harga Asli: Rp{self.harga}")
class Elektronik(Produk):
    def __init__(self, nama, harga, garansi):
        super().__init__(nama, harga)
        self.garansi = garansi
    def diskon(self):
        potongan = self.harga * 0.10
        return self.harga - potongan
    def info(self):
        super().info()
        print(f"Garansi: {self.garansi}")
        print(f"Harga Setelah Diskon: Rp{int(self.diskon())}\n")
class Pakaian(Produk):
    def __init__(self, nama, harga, ukuran):
        super().__init__(nama, harga)
        self.ukuran = ukuran
    def diskon(self):
        potongan = self.harga * 0.20
        return self.harga - potongan
    def info(self):
        super().info()
        print(f"Ukuran: {self.ukuran}")
        print(f"Harga Setelah Diskon: Rp{int(self.diskon())}\n")

tv = Elektronik("TV Pintar Samsung", 5000000, "1 Tahun")
jaket = Pakaian("Jaket Denim", 400000, "XL")

print("=== PRODUK ELEKTRONIK ===")
tv.info()
print("=== PRODUK PAKAIAN ===")
jaket.info()
```

```python
# 1. Buat Parent Class
class Produk:
    def __init__(self, nama, harga, kategori, stok):
        self.nama = nama
        self.harga = harga
        self.kategori = kategori
        self.stok = stok
    def info(self):
        print(f"Nama Barang : {self.nama}")
        print(f"Kategori    : {self.kategori}")
        print(f"Harga       : Rp{self.harga:,}")
        print(f"Sisa Stok   : {self.stok}")

    def beli(self, jumlah_beli):
        print(f"\n>> Transaksi: Membeli {jumlah_beli} unit {self.nama} <<")
        if jumlah_beli <= self.stok:
            total_harga = self.harga * jumlah_beli
            self.stok -= jumlah_beli
            print(f"Berhasil! Total harga yang harus dibayar: Rp{total_harga:,}")
            print(f"Sisa stok {self.nama} sekarang: {self.stok}")
            return total_harga
        else:
            print(f"Gagal! Stok tidak mencukupi. Sisa stok hanya {self.stok}.")
            return 0
class Elektronik(Produk):
    def __init__(self, nama, harga, kategori, stok, garansi):
        super().__init__(nama, harga, kategori, stok)
        # Tambahkan informasi Garansi
        self.garansi = garansi

    def info(self):
        print("=== INFO PRODUK ELEKTRONIK ===")
        super().info()
        print(f"Garansi     : {self.garansi} Bulan\n")


class Pakaian(Produk):
    def __init__(self, nama, harga, kategori, stok, ukuran):
        super().__init__(nama, harga, kategori, stok)
        # Tambahkan informasi Ukuran
        self.ukuran = ukuran

    def info(self):
        print("=== INFO PRODUK PAKAIAN ===")
        super().info()
        print(f"Ukuran      : {self.ukuran}\n")
produk_marketplace = [
    {"nama": "Laptop Gaming", "harga": 15000000, "kategori": "Elektronik", "garansi": 24, "stok": 5},
    {"nama": "Kemeja Flanel", "harga": 200000, "kategori": "Pakaian", "ukuran": "M", "stok": 10},
    {"nama": "Smart TV", "harga": 7000000, "kategori": "Elektronik", "garansi": 12, "stok": 3},
    {"nama": "Celana Jeans", "harga": 300000, "kategori": "Pakaian", "ukuran": "L", "stok": 7}
]
daftar_objek_produk = []
for data in produk_marketplace:
    if data["kategori"] == "Elektronik":
        barang = Elektronik(data["nama"], data["harga"], data["kategori"], data["stok"], data["garansi"])
    elif data["kategori"] == "Pakaian":
        barang = Pakaian(data["nama"], data["harga"], data["kategori"], data["stok"], data["ukuran"])

    daftar_objek_produk.append(barang)


# --- 4. ALUR SISTEM ---

# A. Menampilkan semua barang yang ada di toko
print("--- KATALOG TOKO ---")
for produk in daftar_objek_produk:
    produk.info()

# B. Simulasi User Membeli Produk dan Sistem Menghitung Harga
# Mari kita simulasikan membeli Laptop Gaming (indeks ke-0 di list) dan Kemeja (indeks ke-1)

laptop_gaming = daftar_objek_produk[0]
kemeja_flanel = daftar_objek_produk[1]

laptop_gaming.beli(2)

kemeja_flanel.beli(3)

smart_tv = daftar_objek_produk[2]
smart_tv.beli(10)
```
