```python
class Kendaraan:
  def __init__(self, kecepatan):
    self.kecepatan = kecepatan
  def berjalan(self):
    print("Kendaraan sedang berjalan ")

  def laju(self):
    print (f"dengan kecepatan {self.kecepatan} km/jam")

class Mobil(Kendaraan):
  def berjalan(self):
    print("Mobil berjalan di jalan raya")

  def laju(self):
    print (f"dengan kecepatan {self.kecepatan} km/jam")

class Motor(Kendaraan):
  def berjalan(self):
    print("Motor melaju di jalan raya")

  def laju(self):
    print (f"dengan kecepatan {self.kecepatan} km/jam")

class Pesawat(Kendaraan):
  def berjalan(self):
    print("Pesawat terbang di udara")

  def laju(self):
    print (f"dengan kecepatan {self.kecepatan} km/jam")

class Kapal (Kendaraan):
  def berjalan(self):
    print("Kapal Berlayar di Lautan" )

  def laju(self):
    print (f"dengan kecepatan {self.kecepatan} km/jam")

daftar_kendaraan = [Mobil(100),Motor(50),Pesawat(900),Kapal(100)]

for kendaraan in daftar_kendaraan:
    kendaraan.berjalan()
    kendaraan.laju()
```

```python
class Karyawan:
    def __init__(self, nama):
        self.nama = nama

    def hitung_gaji(self):
        pass

    def gaji_bersih(self):
        gaji_kotor = self.hitung_gaji()
        pajak = 0

        if gaji_kotor > 5000000:
            pajak = 0.05 * gaji_kotor

        return gaji_kotor - pajak

class KaryawanTetap(Karyawan):
    def __init__(self, nama, gaji_bulanan):
        super().__init__(nama)
        self.gaji_bulanan = gaji_bulanan

    def hitung_gaji(self):
        bonus = 0.10 * self.gaji_bulanan
        return self.gaji_bulanan + bonus

class KaryawanFreeLance(Karyawan):
    def __init__(self, nama, jam_kerja, tarif_perjam):
        super().__init__(nama)
        self.tarif_perjam = tarif_perjam
        self.jam_kerja = jam_kerja

    def hitung_gaji(self):
        return self.jam_kerja * self.tarif_perjam

daftar_karyawan = [
    KaryawanTetap("Andi", 5000000),    # Gaji kotor: 5.5jt (karena bonus), kena pajak
    KaryawanFreeLance("Budi", 10, 200000),
    KaryawanTetap("Dimas", 4000000),   # Gaji kotor: 4.4jt, tidak kena pajak
    KaryawanFreeLance("Eka", 40, 150000) # Gaji kotor: 6jt, kena pajak
]

for karyawan in daftar_karyawan:
    print(f"Nama         : {karyawan.nama}")
    print(f"Gaji Kotor   : Rp {karyawan.hitung_gaji():,.0f}")
    print(f"Gaji Bersih  : Rp {karyawan.gaji_bersih():,.0f}")
    print("-" * 30)
```

```python
import json
import os
from datetime import datetime

class Barang:
    def __init__(self, id_barang, nama, harga):
        self.id_barang = id_barang
        self.nama = nama
        self.harga = harga

    def info(self):
        return f"[{self.id_barang}] {self.nama} - Rp{self.harga:,}"

    def to_dict(self):
        return {
            "id_barang": self.id_barang,
            "nama": self.nama,
            "harga": self.harga
        }


class BarangElektronik(Barang):
    def __init__(self, id_barang, nama, harga, garansi):
        super().__init__(id_barang, nama, harga)
        self.garansi = garansi

    def info(self):
        return super().info() + f" | Garansi: {self.garansi} bulan"

    def to_dict(self):
        # Menyimpan identitas tipe kelas
        data = super().to_dict()
        data.update({"tipe": "elektronik", "garansi": self.garansi})
        return data


class BarangKonsumsi(Barang):
    def __init__(self, id_barang, nama, harga, tgl_exp):
        super().__init__(id_barang, nama, harga)
        self.tgl_exp = tgl_exp

    def info(self):
        info_dasar = super().info()
        status_kadaluarsa = ""

        try:
            exp_date = datetime.strptime(self.tgl_exp, "%Y-%m-%d").date()
            hari_ini = datetime.now().date()

            if hari_ini > exp_date:
                status_kadaluarsa = " [❌ KADALUARSA]"
            elif hari_ini == exp_date:
                status_kadaluarsa = " [⚠️ KADALUARSA HARI INI]"
            else:
                status_kadaluarsa = " [✅ AMAN]"
        except ValueError:
            status_kadaluarsa = " [Format Tanggal Error]"

        return info_dasar + f" | Exp: {self.tgl_exp}{status_kadaluarsa}"

    def to_dict(self):
        # Menyimpan identitas tipe kelas
        data = super().to_dict()
        data.update({"tipe": "konsumsi", "tgl_exp": self.tgl_exp})
        return data

class GudangPolimorfik:
    def __init__(self):
        self.daftar_barang = []

    def tambah_barang(self, barang):
        self.daftar_barang.append(barang)

    def tampilkan_barang(self):
        print("\n=== LAPORAN STOK GUDANG ===")
        if not self.daftar_barang:
            print("Gudang saat ini kosong.")
        else:
            for barang in self.daftar_barang:
                print(barang.info())
        print("===========================\n")

    def simpan_data(self, filename="data_gudang.json"):
        data = [barang.to_dict() for barang in self.daftar_barang]
        with open(filename, 'w') as file:
            json.dump(data, file, indent=4)
        print(f"[*] Data berhasil disimpan ke '{filename}'")

    def muat_data(self, filename="data_gudang.json"):
        if os.path.exists(filename):
            with open(filename, 'r') as file:
                data_json = json.load(file)
                self.daftar_barang = []

                for item in data_json:
                    if item.get("tipe") == "elektronik":
                        b = BarangElektronik(item["id_barang"], item["nama"], item["harga"], item["garansi"])
                        self.daftar_barang.append(b)
                    elif item.get("tipe") == "konsumsi":
                        b = BarangKonsumsi(item["id_barang"], item["nama"], item["harga"], item["tgl_exp"])
                        self.daftar_barang.append(b)
            print(f"[*] Data berhasil dimuat dari '{filename}'")
        else:
            print(f"[*] File '{filename}' belum ada, mulai dengan memori kosong.")

def main():
    gudang = GudangPolimorfik()

    print("--- MENAMBAHKAN BARANG KE GUDANG ---")
    gudang.tambah_barang(BarangElektronik("E001", "Laptop ASUS", 12000000, 24))
    gudang.tambah_barang(BarangElektronik("E002", "Kabel Data", 50000, 1))

    gudang.tambah_barang(BarangKonsumsi("K001", "Biskuit Gandum", 25000, "2027-10-15"))
    gudang.tambah_barang(BarangKonsumsi("K002", "Susu Kotak", 8000, "2022-05-10"))

    gudang.tampilkan_barang()

    gudang.simpan_data()

    print("\n--- MENGOSONGKAN MEMORI (Simulasi Restart) ---")
    gudang.daftar_barang = []
    gudang.tampilkan_barang()
    print("--- MEMUAT ULANG DATA DARI JSON ---")
    gudang.muat_data()

    gudang.tampilkan_barang()


if __name__ == "__main__":
    main()
```
